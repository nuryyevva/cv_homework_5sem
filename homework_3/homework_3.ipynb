# Задача №1 - Лес или пустыня?

Часто при анализе изображений местности необходимо понять ее характер. В частности, если определить, что на изображении преобладет вода, то имеет смысл искать корабли на таком изображении. Если на картинке густой лес, то, возможно, это не лучшая зона для посадки дрона или беспилотника.

Ваша задача - написать программу, которая будет отличать лес от пустыни. В приложении можно найти реальные спутниковые снимки лесов и пустынь.

```python
# Load images
forest_images = [
    cv2.imread('desert_forest/test_image_00.jpg'),
    cv2.imread('desert_forest/test_image_02.jpg'),
    cv2.imread('desert_forest/test_image_07.jpg'),
    cv2.imread('desert_forest/test_image_09.jpg'),
]

desert_images = [
    cv2.imread('desert_forest/test_image_04.jpg'),
    cv2.imread('desert_forest/test_image_06.jpg'),
    cv2.imread('desert_forest/test_image_13.jpg'),
    cv2.imread('desert_forest/test_image_15.jpg'),
]

# Calculate histograms
hist_forest = np.zeros((256, 1))
hist_desert = np.zeros((256, 1))

for image in forest_images:
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    hist_forest += cv2.calcHist([gray_image.ravel()], [0], None, [256], [0, 256])

for image in desert_images:
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    hist_desert += cv2.calcHist([gray_image.ravel()], [0], None, [256], [0, 256])

# Normalize histograms
hist_forest /= len(forest_images)
hist_desert /= len(desert_images)
hist_forest /= hist_forest.max()
hist_desert /= hist_desert.max()

def classify_image(image_path, hist_forest, hist_desert):
    """Classifies an image as forest or desert based on histogram comparison."""
    img = cv2.imread(image_path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    hist_img = cv2.calcHist([img_gray.ravel()], [0], None, [256], [0, 256])
    hist_img /= hist_img.max()

    cor_f = cv2.compareHist(hist_forest.astype(np.float32), hist_img, cv2.HISTCMP_CORREL)
    cor_d = cv2.compareHist(hist_desert.astype(np.float32), hist_img, cv2.HISTCMP_CORREL)

    if cor_f > 0.01 and cor_d < -0.01:
        return "Forest"
    elif cor_f < -0.01 and cor_d > 0.01:
        return "Desert"
    else:
        return "Unknown"  # Handle cases where neither threshold is met


# Classify images and store results
desert_forest = []
for i in range(17):
    if i < 10:
        image_path = f'desert_forest/test_image_0{i}.jpg'
    else:
        image_path = f'desert_forest/test_image_{i}.jpg'
    classification = classify_image(image_path, hist_forest, hist_desert)
    desert_forest.append(classification)


plt.figure(figsize=(10, 6))
for i, classification in enumerate(desert_forest):
    if i < 10:
        image_path = f'desert_forest/test_image_0{i}.jpg'
    else:
        image_path = f'desert_forest/test_image_{i}.jpg'
    
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    plt.subplot(3, 6, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f"Image {i+1}: {classification}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Задача №2 - Реализовать Image-blending на основе сшивки по градиентам

Задача - взять фото двух лиц : ваше и друга, с помощью метода Poisson image editing совместить глаза, нос и рот с первого изображения со вторым. Суть в том, что при использовании такого метода границы совмещенного изображения не видны.

Статья, где описан метод

```python
# Load the source and target
source_image = cv2.imread('blending/zendaya.jpg')
source_image = cv2.cvtColor(source_image, cv2.COLOR_BGR2RGB)
target_image = cv2.imread('blending/me.jpg')
target_image = cv2.cvtColor(target_image, cv2.COLOR_BGR2RGB)
# Load mask image
mask = cv2.imread('blending/mask.jpg', 0)

height, width, channels = target_image.shape

normal_clone = cv2.seamlessClone(source_image, target_image, mask, (width//2-82, height//2-8), cv2.NORMAL_CLONE)

plt.figure(figsize=(12, 6), constrained_layout=False)
plt.subplot(1, 3, 1)
plt.imshow(source_image)
plt.title('Source Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(normal_clone)
plt.title('Result Image')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(target_image)
plt.title('Target Image')
plt.axis('off')
plt.show()
```

```python
def laplacian(image):
    kernel = np.array([[0, 1, 0],
                       [1, -4, 1],
                       [0, 1, 0]])
    return cv2.filter2D(image, -1, kernel)

def solve_poisson(f, mask, regularization=1e-3):
    h, w = f.shape[:2]
    n = h * w

    # Create a sparse matrix using dok_matrix 
    A = dok_matrix((n, n), dtype=np.float32)

    mask_flat = mask.flatten()

    # Add values for neighboring pixels 
    for i in range(n):
        row = i // w
        col = i % w

        # Add values for neighboring pixels 
        if row > 0:
            A[i, i - w] = 1  # Up
        if row < h - 1:
            A[i, i + w] = 1  # Down
        if col > 0:
            A[i, i - 1] = 1  # Left
        if col < w - 1:
            A[i, i + 1] = 1  # Right

        # Set diagonal value
        A[i, i] = -4

    # Convert to csc_matrix 
    A = A.tocsc()

    # Apply regularization (add small value to the diagonal elements directly)
    A.data[A.indptr[:-1]] += regularization

    f_flat = f.flatten()

    # Solve the linear system
    u_flat = spsolve(A, f_flat)

    u = u_flat.reshape(h, w)
    return u

def poisson_blend(source, target, mask):
    blended_image = np.zeros_like(source)
    for channel in range(3):
        source_region = source[:, :, channel] * mask
        target_region = target[:, :, channel] * (1 - mask)

        f = laplacian(source_region) - laplacian(target_region)
        u = solve_poisson(f, mask)

        blended_image[:, :, channel] = target_region + u * mask
        # plot_transform_result(f, cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB), is_gray=True)
    
    return blended_image

source_image = cv2.imread("blending/zendaya.jpg")
target_image = cv2.imread("blending/me.jpg")
mask_image = cv2.imread("blending/mask.jpg", cv2.IMREAD_GRAYSCALE)
source_image = cv2.resize(source_image, None, fx=0.25, fy=0.25)
target_image = cv2.resize(target_image, None, fx=0.25, fy=0.25)
mask_image = cv2.resize(mask_image, None, fx=0.25, fy=0.25)
mask_image = mask_image // 255.0  # Mask as a binary image

blended_image = poisson_blend(source_image, target_image, mask_image)

# Display the blended image
plt.figure(figsize=(16, 12), constrained_layout=False)
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(source_image, cv2.COLOR_BGR2RGB))
plt.title('Source Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB))
plt.title('Result Image')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(target_image, cv2.COLOR_BGR2RGB))
plt.title('Target Image')
plt.axis('off')
plt.show()
```

# Задача №3 - Найди клетки

Даны снимки раковых клеток. Существует задача - определить стадию рака клетки по такому изображению. Для того, чтобы подойти к решению классификации рака клетки, необходимо сначала подготовить данные.

Исходные изображения в реальных задачах могут быть очень большого размера (более 20000 px). Однако из визуального анализа можно заметить, что большая часть этих снимков пустая и не несет в себе полезную информацию.

Ваша задача выделить небольшие ячейки изображений из исходного так, чтобы на ячейках было только изображение клетки.

В качестве аргументов у функции будут значения:
1. исходное изображние;
2. размер ячейки;
3. количество ячеек.

```python
def extract_cells(image, cell_size, cell_count):
    """Extracts cells from an image."""

    # Convert to grayscale and binarize
    gray_image = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2GRAY)
    _, binary_image = cv2.threshold(gray_image, 254, 1, cv2.THRESH_BINARY_INV)

    # Create a kernel for convolution
    kernel = np.ones((cell_size, cell_size))

    # Find count of cell pixels in every possible kernel position
    convolution = cv2.filter2D(binary_image.astype(np.float32), -1, kernel)

    # Apply zero padding to avoid edge effects
    padding = cell_size // 2
    convolution[:padding, ...] = 0
    convolution[-padding:, ...] = 0
    convolution[..., :padding] = 0
    convolution[..., -padding:] = 0

    # Extract cell squares
    extracted_cells = []
    shift = cell_size // 2
    move = (cell_size + 1) % 2  # Adjust shift for even kernel sizes
    for _ in range(cell_count):
        if not np.any(convolution):  # Break if no more possible squares
            break
        y, x = np.unravel_index(convolution.argmax(), convolution.shape)
        cell = image[y - shift:y + shift - move + 1, x - shift:x + shift - move + 1]
        extracted_cells.append(cell)

        # Clear area of the square and a buffer around it
        top_bound = max(0, y - 2 * shift)
        bottom_bound = min(convolution.shape[0], y + 2 * (shift - move) + 1)
        left_bound = max(0, x - 2 * shift)
        right_bound = min(convolution.shape[1], x + 2 * (shift - move) + 1)
        convolution[top_bound:bottom_bound, left_bound:right_bound] = 0

    # Create the final image with extracted cells
    width = int(np.sqrt(cell_count))
    height = cell_count // width + cell_count % width
    if len(extracted_cells) < width * height:
        # Fill in missing cells with white squares if there aren't enough
        empty_image_count = width * height - len(extracted_cells)
        for _ in range(empty_image_count):
            extracted_cells.append(np.full((cell_size, cell_size), 255))

    final_image = np.zeros((cell_size, width * cell_size, 3), dtype=np.uint8)  # Add 3rd dimension for RGB
    for i in range(height):
        line = np.hstack(tuple(extracted_cells[i * width:(i + 1) * width]))
        final_image = np.vstack((final_image, line))

    final_image = final_image[cell_size:, ...]  # Remove initial row

    return final_image

# Example usage
image = cv2.imread("cells/train3_1.jpeg")
res = extract_cells(image, 350, 16)

# Display results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
plt.title('Extracted Cells')
plt.axis('off')
plt.show()
```
