````python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def plot_one_image(image: np.ndarray) -> None:
    """
    Отобразить изображение с помощью matplotlib.
    Вспомогательная функция.
    
    :param image: изображение для отображения
    :return: None
    """
    fig, axs = plt.subplots(1, 1, figsize=(8, 7))
    
    axs.imshow(image)
    axs.axis('off')
    plt.plot()
````

Реализуйте алгоритм поиска выхода из лабиринта по растровому изобажению.

Отрисуйте получившийся маршрут на карте с помощью функции ```plot_maze_path(img, coords)``` или воспользуйтесь вам известным графическим инструментом.

__Input:__

Все карты лежат на [яндекс-диске](https://yadi.sk/d/qEWVZk2picDdZw)

__Ouput:__

Массив координат пути через лабиринт в виде ```(np.array(x), np.array(y))```. 

Оценивается __каждое__ успешное решение лабиринта.
Пример решенной задачи.<img src="https://b.radikal.ru/b22/1906/56/df991f2221a7.png" width="300">

````python
from task_11 import find_way_from_maze
````

````python
def plot_maze_path(image: np.ndarray, coords: tuple) -> np.ndarray:
    """
    Нарисовать путь через лабиринт на изображении. 
    Вспомогательная функция.
     
    :param image: изображение лабиринта
    :param coords: координаты пути через лабиринт типа (x, y) где x и y - массивы координат точек
    :return img_wpath: исходное изображение с отрисованными координатами 
    """
    if image.ndim != 3:
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        
    img_wpath = image.copy()
    if coords:
        x, y = coords
        img_wpath[x, y, :] = [0, 0, 255]

    return img_wpath
````

Загрузим тестовое изображение и отобразим его.

````python
test_image = cv2.imread('task_1/20 by 20 orthogonal maze.png')  # загрузить тестовую картинку

plot_one_image(test_image)
````

Теперь ваша задача реализовать функцию ```plot_maze_path``` в ```task_1.py``` для того, чтобы найти координаты пути через лабиринт.

````python
way_coords = find_way_from_maze(test_image)  # вычислить координаты пути через лабиринт

image_with_way = plot_maze_path(test_image, way_coords)

plot_one_image(image_with_way)
````

## Задача №3.2 - Афинные преобразования

Реализуйте прототип программы, которая помогает сканировать документы с помощью приложения на телефоне.

<img src="task_3/adobe_scan.jpg"/>

Данная программа берёт изображение, находит углы листа и приводит документ к виду с вертикальной проекцией с помощью афинных преобразований.

В данной задаче предлагается обработать фото тетради, найдя её углы с помощью поиска по цвету.

````python
image = cv2.imread('task_3/notebook.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_mask = cv2.imread('task_3/notebook_mask.jpeg')
image_mask = cv2.cvtColor(image_mask, cv2.COLOR_BGR2RGB)

image_points = cv2.imread('task_3/notebook_points.png')
image_points = cv2.cvtColor(image_points, cv2.COLOR_BGR2RGB)

image_result = cv2.imread('task_3/notebook_result.png')
image_result = cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')    

plt.subplot(2, 2, 2)
plt.title("Image mask")
plt.imshow(image_mask, cmap="gray")
plt.axis('off')  

plt.subplot(2, 2, 3)
plt.title("Image with points")
plt.imshow(image_points)
plt.axis('off')  

plt.subplot(2, 2, 4)
plt.title("Result Image")
plt.imshow(image_result)
plt.axis('off')  

plt.show()
````

````python
notebook_image  = cv2.imread('task_3/notebook.jpg')
notebook_image = cv2.cvtColor(notebook_image, cv2.COLOR_BGR2RGB)  # convert into RGB
notebook_image_hsv = cv2.cvtColor(notebook_image, cv2.COLOR_RGB2HSV) # convert into HSV

# create mask of red area
lower_red1 = np.array([0, 120, 70])
upper_red1 = np.array([5, 255, 255])
lower_red2 = np.array([175, 120, 70])
upper_red2 = np.array([180, 255, 255])

mask1 = cv2.inRange(notebook_image_hsv, lower_red1, upper_red1)
mask2 = cv2.inRange(notebook_image_hsv, lower_red2, upper_red2)

notebook_area = mask1 + mask2

# find corners of the notebook and draw circles
indices = np.where(notebook_area != 0) # Find non-zero pixels in the mask
y_coords = indices[0] # Extract y coordinates of non-zero pixels
x_coords = indices[1] # Extract x coordinates of non-zero pixels
# find extreme points based on coordinates
most_upper_point = (x_coords[np.argmin(y_coords)], np.min(y_coords))
most_lower_point = (x_coords[np.argmax(y_coords)], np.max(y_coords))
most_left_point = (np.min(x_coords), y_coords[np.argmin(x_coords)])
most_right_point = (np.max(x_coords), y_coords[np.argmax(x_coords)])
image_with_points = notebook_image.copy()
# draw
cv2.circle(image_with_points, most_upper_point, 10, (0, 0, 255), -1)  # Blue
cv2.circle(image_with_points, most_lower_point, 10, (0, 255, 0), -1)  # Green
cv2.circle(image_with_points, most_left_point, 10, (255, 0, 0), -1)  # Red
cv2.circle(image_with_points, most_right_point, 10, (255, 165, 0), -1)  # Orange

# do projective transformation
rows, cols, _ = notebook_image.shape
pts1 = np.float32([most_upper_point, most_lower_point, most_left_point, most_right_point])
pts2 = np.float32([[0, 0],[cols, rows],[0, rows], [cols, 0]]) 
M = cv2.getPerspectiveTransform(pts1, pts2) 
result = cv2.warpPerspective(image, M, (cols, rows))

# plot the result
fig, m_axs = plt.subplots(1, 4, figsize = (15, 10))
ax1, ax2, ax3, ax4 = m_axs

ax1.set_title('Original image', fontsize=11)
ax1.imshow(notebook_image)
ax2.set_title('Mask', fontsize=11)
ax2.imshow(notebook_area, cmap='gray');
ax3.set_title('Notebook with points', fontsize=11)
ax3.imshow(image_with_points, cmap='gray');
ax4.set_title('Only notebook', fontsize=11)
ax4.imshow(result, cmap='gray');
````
