```python
import cv2
import numpy as np
from sklearn.feature_extraction.image import grid_to_graph
from sklearn.cluster import AgglomerativeClustering
import skimage.io
import matplotlib.pyplot as plt
import os
from scipy import ndimage as ndi
from skimage.segmentation import watershed
from skimage.feature import peak_local_max
```

```python
# вспомогательная функция
def plot_one_image(src_image, is_gray=False):
    """
    Отрисовать с помощью plt исходное изображение.
    
    :param src_image: np.ndarray: исходное изображение
    :param is_gray: bool: флаг для отображения ЧБ изображений
    :return: None
    """
    fig, m_axs = plt.subplots(1, 1, figsize=(6.4, 4.8), constrained_layout=False)
    ax1 = m_axs

    cmap = 'gray' if is_gray else None
    ax1.set_title('Исходное изображение')
    ax1.imshow(src_image, cmap=cmap)
    ax1.set_xticks([]), ax1.set_yticks([])
    plt.show()
```

```python
# вспомогательная функция
def plot_transform_result(src_image, transform_image, is_gray=False, name_1='Исходное изображение', name_2='Результат преобразования', figsize=(6.4*2, 4.8*2)):
    """
    Отрисовать с помощью plt исходное изображение и его преобразование.
    
    :param src_image: np.ndarray: исходное изображение
    :param transform_image: np.ndarray: преобразованное изображение
    :param is_gray: bool: флаг для отображения ЧБ изображений
    :return: None
    """
    fig, m_axs = plt.subplots(1, 2, figsize=figsize, constrained_layout=False)
    ax1, ax2 = m_axs

    cmap = 'gray' if is_gray else None
    ax1.set_title(name_1)
    ax1.imshow(src_image, cmap=cmap)
    ax1.set_xticks([]), ax1.set_yticks([])
    ax2.set_title(name_2)
    ax2.imshow(transform_image, cmap=cmap)
    ax2.set_xticks([]), ax2.set_yticks([])
    plt.show()
```

```python
def plot_all_images(lst, row, col):
    plt.figure(figsize=(14, 14), constrained_layout=True)
    i = 0
    for img in lst:
        #img = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)
        plt.subplot(row, col, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f'Image: {i}')
        plt.axis('off')
        i += 1
    plt.show()

def invert_black_white(images):
    lst = []
    for img in images:
        #image = cv2.subtract(255, img.copy())
        image = cv2.bitwise_not(img.copy())
        lst.append(image)
    return lst
```

```python
def load_images_from_folder(folder1, folder2):
    images1, images2 = [], []
    for filename in os.listdir(folder1):
        if filename[-3:] != 'txt':
            img1 = skimage.io.imread(os.path.join(folder1,filename))
            filename = f"{filename[:-3]}png"
            img2 = skimage.io.imread(os.path.join(folder2,filename))
            if img1 is not None:
                images1.append(img1)
            if img2 is not None:
                images2.append(img2)
    return images1, images2
ships, ships_gt = load_images_from_folder("ships/img", "ships/ground_truth")
```

```python
def agglomerative_clustering(folder, k):
    lst = []
    i = 0
    for image in folder:
        image = skimage.transform.resize(image.copy(), tuple(map(int, np.array(image.shape[:2])/k)))
        X = np.reshape(image, (-1, 1))
    
        # Define the structure A of the data. Pixels connected to their neighbors.
        connectivity = grid_to_graph(*image.shape)

        # Compute clustering
        print(str(i)+":", "Compute structured hierarchical clustering...")
        ward = AgglomerativeClustering(n_clusters=2, memory = "cache", connectivity=connectivity,
                linkage='ward', distance_threshold=None).fit(X)

        label_aggl = np.reshape(ward.labels_, image.shape)
        res = label_aggl[...,0]
        lst.append(res)
        i+=1
    return lst
```

# Задача №2 - Отделить три класса (треугольник, круг, квадрат) друг от друга

Реализуйте программу, которая делит изображение на три класса : треугольник, круг, квадрат. Сложность состоит в том, что сегментацию нужно выполнить и на зашумленных данных. Все картинки лежат в папке figures/img, разметка к ним - в figures/gt

Примеры изображений:
<table><tr>
    <td> <img src="figures/img/little_noise.png" alt="Drawing" style="width: 400px;"/> </td>
    <td> <img src="figures/gt/little_noise.png" alt="Drawing" style="width: 400px;"/> </td>
</tr></table>

```python
def get_black_and_white_image(images):
    lst = []
    for image in images:
        kernel = np.ones((5, 5), np.uint8)
        img = image.copy()
        gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

        blur = cv2.GaussianBlur(gray_img.copy(), (13,13), 0).astype(np.uint16)
        ret, th = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        opening = cv2.morphologyEx(th, cv2.MORPH_OPEN, kernel)
        closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
        
        laplac = cv2.Laplacian(closing, cv2.THRESH_BINARY, scale=1, ksize=5)
        contours, hierarchy = cv2.findContours(laplac, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        new_contours = []
        for cnt in contours:
            if cnt.shape[0] > 500 and cnt.shape[0] < 600:
                new_contours.append(cnt)

        # smooth contour
        peri = cv2.arcLength(new_contours[0], True)
        new_contours[0] = cv2.approxPolyDP(new_contours[0].copy(), 0.003 * peri, True)

        contour_img = np.zeros(shape=gray_img.shape)
        cv2.drawContours(contour_img, new_contours, 0, (255, 0, 0), -1)
        
        lst.append(contour_img)
    return lst
```

```python
def colour(images):
    lst = []
    for image in images:
        image = image.astype(np.int32)
        distance = ndi.distance_transform_edt(image)
        coords = peak_local_max(distance, footprint=np.ones((5, 5)), labels=image)
        mask = np.zeros(distance.shape, dtype=bool)
        mask[tuple(coords.T)] = True
        markers, _ = ndi.label(mask)
        labels = watershed(-distance, markers, mask=image, compactness=0.5)
        img = labels.copy()
        img[ img > 3 ] = 15
        img[ img  == 1 ] = 3
        img[ img  == 2 ] = 3
        lst.append(img)
    return lst
```

```python
figures = []
for filename in os.listdir("figures/img"):
    img = skimage.io.imread(os.path.join("figures/img",filename))
    if img is not None:
        figures.append(img)
plot_all_images(figures, 1, 3)
black_and_white_images = get_black_and_white_image(figures)
# plot_all_images(black_and_white_images, 1, 3)
coloured = colour(black_and_white_images)
plot_all_images(coloured, 1, 3)
```
