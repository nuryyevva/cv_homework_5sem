```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from pathlib import Path
```

```python
# вспомогательная функция
def plot_one_image(src_image, is_gray=False):
    """
    Отрисовать с помощью plt исходное изображение.
    
    :param src_image: np.ndarray: исходное изображение
    :param is_gray: bool: флаг для отображения ЧБ изображений
    :return: None
    """
    fig, m_axs = plt.subplots(1, 1, figsize=(6, 4), constrained_layout=True)
    ax1 = m_axs

    cmap = 'gray' if is_gray else None
    ax1.set_title('Исходное изображение')
    ax1.imshow(src_image, cmap=cmap)
    ax1.set_xticks([]), ax1.set_yticks([])
    plt.show()
```

```python
def load_images_from_folder(folder):    
    images = []
    for filename in os.listdir(folder):        
        img = cv2.imread(os.path.join(folder,filename))
        img = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)
        if img is not None:            
            images.append(img)
    return images
```

```python
low_green = (30, 0, 0)
high_green = (90, 255, 255)
def get_only_leaves(lst):
    masked_lst = []
    for img in lst:
        img_cp = cv2.cvtColor(img.copy(), cv2.COLOR_RGB2HSV)
        mask = cv2.inRange(img_cp, low_green, high_green)
        x, y = np.argwhere(mask == 255).T
        only_leaves = np.zeros_like(img)
        only_leaves[x, y] = img[x, y]
        masked_lst.append(only_leaves)
    return masked_lst
```

```python
def count_of_good_matches(test_image, class_images):
    count = 0
    for image in class_images:
        # Initiate SIFT detector 
        sift = cv2.SIFT_create()

        # Detect and compute the keypoints and descriptors with SIFT 
        kp1, des1 = sift.detectAndCompute(test_image, None) 
        kp2, des2 = sift.detectAndCompute(image, None)

        # Create BFMatcher object 
        bf = cv2.BFMatcher()

        # Match descriptors. 
        matches = bf.match(des1, des2)

        # Sort the matches based on distance 
        matches = sorted(matches, key=lambda val: val.distance)

        count += sum([x.distance for x in matches[:200]])

    return count
```

```python
# Load images
loose = load_images_from_folder("plants/train/Loose Silky-bent")
maize = load_images_from_folder("plants/train/Maize")
mayweed = load_images_from_folder("plants/train/Scentless Mayweed")
cranesbill = load_images_from_folder("plants/train/Small-flowered Cranesbill")

loose_masked = get_only_leaves(loose)
maize_masked = get_only_leaves(maize)
mayweed_masked = get_only_leaves(mayweed)
cranesbill_masked = get_only_leaves(cranesbill)

tests = load_images_from_folder("plants/test")
tests_masked = get_only_leaves(tests)
```

```python
# Find best matches
classes = []
for i in range(len(tests)):
    loose_number = count_of_good_matches(tests_masked[i].copy(), loose_masked.copy())
    maize_number = count_of_good_matches(tests_masked[i].copy(), maize_masked.copy())
    mayweed_number = count_of_good_matches(tests_masked[i].copy(), mayweed_masked.copy())
    cranesbill_number = count_of_good_matches(tests_masked[i].copy(), cranesbill_masked.copy())
    nums = [loose_number, maize_number, mayweed_number, cranesbill_number]
    classes.append(nums.index(min(nums)))

classes = ['Loose Silky-bent' if x==0 else x for x in classes]
classes = ['Maize' if x==1 else x for x in classes]
classes = ['Scentless Mayweed' if x==2 else x for x in classes]
classes = ['Small-flowered cranesbill' if x==3 else x for x in classes]
```

```python
# Plot
plt.figure(figsize=(15, 10), constrained_layout=True)
i = 0
for img in tests:
    plt.subplot(5, 8, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(classes[i])
    plt.axis('off')
    i += 1
plt.show()
```

```python
def return_metrics(gt, pred):
    gt = np.array(gt) if isinstance(gt, list) else gt
    pred = np.array(pred) if isinstance(pred, list) else pred

    classes = np.unique(np.concatenate((gt, pred)))
    class_map = {c: i for i, c in enumerate(classes)}
    num_classes = len(classes)

    gt_numerical = np.array([class_map[label] for label in gt])
    pred_numerical = np.array([class_map[label] for label in pred])

    for i in range(num_classes):
        is_class_i_true = (gt_numerical == i)
        is_class_i_pred = (pred_numerical == i)

        true_positives = np.sum(is_class_i_true & is_class_i_pred)
        true_negatives = np.sum(~is_class_i_true & ~is_class_i_pred)

        accuracy = (true_positives + true_negatives) / len(gt_numerical) if len(gt_numerical) > 0 else 0
        recall = true_positives / np.sum(is_class_i_true) if np.sum(is_class_i_true) > 0 else 0

        print(f'Class: {classes[i]}:')
        print(f'  Acc: {accuracy:.2f}')
        print(f'  Recall: {recall:.2f}')
```

```python
# Calculate metrics
gt = ['Scentless Mayweed', 'Small-flowered cranesbill', 'Small-flowered cranesbill', 'Loose Silky-bent', 'Scentless Mayweed', 'Loose Silky-bent', 'Small-flowered cranesbill', 'Scentless Mayweed', 'Maize', 'Maize', 'Loose Silky-bent', 'Small-flowered cranesbill', 'Maize', 'Loose Silky-bent', 'Loose Silky-bent', 'Maize', 'Loose Silky-bent', 'Scentless Mayweed', 'Small-flowered cranesbill', 'Scentless Mayweed', 'Maize', 'Small-flowered cranesbill', 'Maize', 'Small-flowered cranesbill', 'Loose Silky-bent', 'Scentless Mayweed', 'Maize', 'Loose Silky-bent', 'Scentless Mayweed', 'Scentless Mayweed', 'Scentless Mayweed', 'Maize', 'Loose Silky-bent', 'Maize', 'Small-flowered cranesbill', 'Small-flowered cranesbill', 'Maize', 'Scentless Mayweed', 'Loose Silky-bent', 'Small-flowered cranesbill']
return_metrics(gt, classes)
```

```python
def trim(img):
    #crop top
    if not np.sum(img[0]):
        return trim(img[1:])
    #crop bottom
    if not np.sum(img[-1]):
        return trim(img[:-2])
    #crop left
    if not np.sum(img[:,0]):
        return trim(img[:,1:])
    #crop right
    if not np.sum(img[:,-1]):
        return trim(img[:,:-2])
    return img

def find_matches(img1, img2):
    img1, img2 = img1.copy(), img2.copy()
    img1_gray = cv2.cvtColor(img1.copy(), cv2.COLOR_RGB2GRAY)
    img2_gray = cv2.cvtColor(img2.copy(), cv2.COLOR_RGB2GRAY)

    sift = cv2.SIFT_create()
    # find the key points and descriptors with SIFT
    kp1, des1 = sift.detectAndCompute(img1_gray,None)
    kp2, des2 = sift.detectAndCompute(img2_gray,None)
    
    # orb = cv2.ORB_create()
    # kp1 = orb.detect(img1_gray,None)
    # kp1, des1 = orb.compute(img1_gray, kp1)
    # kp2 = orb.detect(img2_gray,None)
    # kp2, des2 = orb.compute(img2_gray, kp2)
    
    if des1 is None or des2 is None:
        print("No keypoints found!")
        return None

    # init match algo
    # FLANN_INDEX_KDTREE = 2
    # index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
    # search_params = dict(checks=50)

    # match = cv2.FlannBasedMatcher(index_params, search_params)
    match = cv2.BFMatcher()
    matches = match.knnMatch(des1, des2, k=2)

    good = []
    for m,n in matches:
        if m.distance < 0.6*n.distance:
            good.append(m)

    # draw_params = dict(matchColor = (0,255,0), singlePointColor = None, flags = 2)
    # img3 = cv2.drawMatches(img1, kp1, img2, kp2, good, None,**draw_params)
    # plot_one_image(img3)
    return good, kp1, kp2
    
def stitch(img1, img2):    
    img1, img2 = img1.copy(), img2.copy()
    res1 = find_matches(img1, img2)
    res2 = find_matches(img2, img1)
    if res1 and res2:
        good1, kp11, kp12 = res1
        good2, kp21, kp22 = res2
    else:
        return []
    
    img1_gray = cv2.cvtColor(img1.copy(), cv2.COLOR_RGB2GRAY)
    img2_gray = cv2.cvtColor(img2.copy(), cv2.COLOR_RGB2GRAY)
    
    MIN_MATCH_COUNT = 4
    if len(good1) >= MIN_MATCH_COUNT:
        src_pts1 = np.float32([ kp11[m.queryIdx].pt for m in good1 ]).reshape(-1,1,2)
        dst_pts1 = np.float32([ kp12[m.trainIdx].pt for m in good1 ]).reshape(-1,1,2)

        M1, mask1 = cv2.findHomography(src_pts1, dst_pts1, cv2.RANSAC, 5.0)
    else:
        return []
    if len(good2) >= MIN_MATCH_COUNT:
        src_pts2 = np.float32([ kp21[m.queryIdx].pt for m in good2 ]).reshape(-1,1,2)
        dst_pts2 = np.float32([ kp22[m.trainIdx].pt for m in good2 ]).reshape(-1,1,2)

        M2, mask2 = cv2.findHomography(src_pts2, dst_pts2, cv2.RANSAC, 5.0)
    else:
        return []
    
    dst1 = cv2.warpPerspective(img1, M1, (img1_gray.shape[1] + img2_gray.shape[1], img1_gray.shape[0] + img2_gray.shape[0]))
    dst2 = cv2.warpPerspective(img2, M2, (img1_gray.shape[1] + img2_gray.shape[1], img1_gray.shape[0] + img2_gray.shape[0]))
    
    dst1[0:img2.shape[0],0:img2.shape[1]] = img2
    dst2[0:img1.shape[0],0:img1.shape[1]] = img1
    if np.sum(dst1 == 0) < np.sum(dst2 == 0):
        return trim(dst1)
    else:
        return trim(dst2)
```

```python
fighter = load_images_from_folder("puzzle/su_fighter_shuffle")
queue = fighter.copy()
temp = fighter[0]
i = 0
while(len(queue) != 0 and i < len(fighter)*2):
    i += 1
    tmp = temp.copy()
    img = queue.pop()
    temp = stitch(temp, img)
    if len(temp) == 0:
        queue.insert(0, img)
        temp = tmp.copy()
    
plot_one_image(temp)
```
