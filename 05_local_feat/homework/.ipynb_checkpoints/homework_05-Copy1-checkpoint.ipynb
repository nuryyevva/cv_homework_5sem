```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# вспомогательная функция
def plot_one_image(src_image, is_gray=True):
    """
    Отрисовать с помощью plt исходное изображение.
    
    :param src_image: np.ndarray: исходное изображение
    :param is_gray: bool: флаг для отображения ЧБ изображений
    :return: None
    """
    fig, m_axs = plt.subplots(1, 1, figsize=(6.4*2, 4.8*2), constrained_layout=True)
    ax1 = m_axs

    cmap = 'gray' if is_gray else None
    ax1.set_title('Исходное изображение')
    ax1.imshow(src_image, cmap=cmap)
    ax1.set_xticks([]), ax1.set_yticks([])
    plt.show()
```

```python
def plot_all_images(lst, row, col):
    plt.figure(figsize=(14, 14), constrained_layout=True)
    i = 0
    for img in lst:
        #img = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)
        plt.subplot(row, col, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f'Image: {i}')
        plt.axis('off')
        i += 1
    
plt.show()
```

# Задача №1 - Можете ли вы отличить сорняки от рассады?

Теперь приступим к задаче классификации на картинках. Реализуйте программу, которая определяет тип рассады на изображении. 

Для того, чтобы определить характерные особенности каждого типа рассады, у вас есть train. Train это папка, в которой картинки уже классифицированы и лежат в соответствующих папках. Исходя из этой информации можете найти признаки, присущие конкретному растению.

```python
low_green = (27, 0, 0)
high_green = (80, 255, 255)
def find_only_leaves(img):
    # img = cv2.cvtColor(img.copy(), cv2.COLOR_GRAY2BGR)
    img = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2HSV)
    # plot_one_image(img)
    leaves_area = cv2.inRange(img.copy(), low_green, high_green)
    # leaves_area = cv2.cvtColor(leaves_area.copy(), cv2.COLOR_HSV2BGR)
    # return cv2.cvtColor(leaves_area.copy(), cv2.COLOR_BGR2GRAY)
    # plot_one_image(leaves_area)
    return leaves_area
    
def get_only_leaves(img):
    mask = find_only_leaves(img)
    x, y = np.argwhere(mask == 255).T
    only_leaves = np.zeros_like(img)
    only_leaves[x, y] = img[x, y]    
    # only_leaves = cv2.cvtColor(only_leaves.copy(), cv2.COLOR_HSV2BGR)
    # return cv2.cvtColor(only_leaves.copy(), cv2.COLOR_BGR2GRAY)
    return only_leaves
```

```python
sift = cv2.SIFT_create()
def extract_sift_features(image):
    keypoints, descriptors = sift.detectAndCompute(image, None)
    return descriptors

train_descriptors = np.empty((0, 128))
train_labels = []
train_dir = "/home/nur/Documents/cv_mipt_major (copy)/05_local_feat/homework/plants/train"

for class_name in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_name)
    if os.path.isdir(class_path):
    
        for image_name in os.listdir(class_path):
            image_path = os.path.join(class_path, image_name)
            # image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE) 
            image = cv2.imread(image_path) 
            # plot_one_image(image)
            masked_image = get_only_leaves(image)
            # plot_one_image(masked_image)
            descriptors = extract_sift_features(masked_image)

            train_descriptors = np.vstack((train_descriptors, descriptors))
            train_labels += [class_name] * len(descriptors)

knn = KNeighborsClassifier(n_neighbors=100)

knn.fit(train_descriptors, train_labels)

test_descriptors = np.empty((0, 128))
test_labels = []

test_dir = "/home/nur/Documents/cv_mipt_major (copy)/05_local_feat/homework/plants/test" 
test_images = []
indices = []
for image_name in os.listdir(test_dir):
    image_path = os.path.join(test_dir, image_name)
    image = cv2.imread(image_path)
    test_images.append(image)
    masked_image = get_only_leaves(image)
    # plot_one_image(masked_image)
    descriptors = extract_sift_features(masked_image)
    indices.append(len(descriptors))
    test_descriptors =  np.vstack((test_descriptors, descriptors))
```

```python
predictions = knn.predict(test_descriptors)
print(len(predictions))
```

```python
sliced_arrays = []
start_index = 0
for end_index in indices:
    sliced_arrays.append(predictions[start_index:start_index+end_index])
    start_index = end_index
classes = []
for u in sliced_arrays:
    # print(len(u))
    unique_elements, counts = np.unique(u, return_counts=True)
    max_index = np.argmax(counts)
    most_common_element = unique_elements[max_index]
    most_common_count = counts[max_index]
    classes.append(most_common_element)
    # print(most_common_element)
plt.figure(figsize=(15, 15), constrained_layout=True)
i=0
for img in test_images:
    plt.subplot(5, 8, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(classes[i])
    plt.axis('off')
    i += 1
plt.show()
```
