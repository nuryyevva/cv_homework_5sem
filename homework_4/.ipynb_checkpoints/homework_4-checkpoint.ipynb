````python
# вспомогательная функция
def plot_transform_result(src_image, transform_image, is_gray=False):
    """
    Отрисовать с помощью plt исходное изображение и его преобразование.
    
    :param src_image: np.ndarray: исходное изображение
    :param transform_image: np.ndarray: преобразованное изображение
    :param is_gray: bool: флаг для отображения ЧБ изображений
    :return: None
    """
    fig, m_axs = plt.subplots(1, 2, figsize=(6.4*2, 4.8*2), constrained_layout=True)
    ax1, ax2 = m_axs

    cmap = 'gray' if is_gray else None
    ax1.set_title('Исходное изображение')
    ax1.imshow(src_image, cmap=cmap)
    ax1.set_xticks([]), ax1.set_yticks([])
    ax2.set_title('Результат преобразования')
    ax2.imshow(transform_image, cmap=cmap)
    ax2.set_xticks([]), ax2.set_yticks([])
    plt.show()
````

````python
# вспомогательная функция
def plot_one_image(src_image, is_gray=False):
    """
    Отрисовать с помощью plt исходное изображение.
    
    :param src_image: np.ndarray: исходное изображение
    :param is_gray: bool: флаг для отображения ЧБ изображений
    :return: None
    """
    fig, m_axs = plt.subplots(1, 1, figsize=(6.4*2, 4.8*2), constrained_layout=True)
    ax1 = m_axs

    cmap = 'gray' if is_gray else None
    ax1.set_title('Исходное изображение')
    ax1.imshow(src_image, cmap=cmap)
    ax1.set_xticks([]), ax1.set_yticks([])
    plt.show()
````

# Задача №1 - Лего детальки

Имеется 2 выборки из датасета фотографий лего картинок. Структура одной выборки следующая:
```
test
| -- 'target.jpg'
| -- source
|    | -- '*.jpg'
|           ...
|    | -- '*.jpg'
```
В выборке есть опорное изображение - target.jpg и набор из 20 картинок в папке source.

Ваша задача - написать программу, которая прочитает все изображения и определит на каких изображениях из папки source сфотографирована такая же лего деталька, что и деталь в target.jpg

Результатом выполнения программы должен быть файлик с названиями подходящих изображений. Пример в Datasets/gt
Снизу представлены 2 вспомогательные функции: чтение всех изображений из папки и посчет получившихся метрик точности

Результат выполнения данной задачи напрямую зависит от полученных метрик. Полученные метрики будут вычисляться из метрик по двум тестам $Accuracy=min({Accuracy}_{test1}, {Accuracy}_{test2}$)

| Accuracy | Recall | % Выполнения |
|:--------:|:------:|:------------:|
|< 0.6|< 0.4|0%|
|  >= 0.6  | >= 0.4 |     75%      |
|  >= 0.7  | >= 0.5 |     100%     |
|  >= 0.9  | >= 0.8 |     150%     |

````python
def read_all_image_paths(folder):
    folder = Path(folder)
    image_paths = [i.__str__() for i in folder.glob('*.jpg')]
    return image_paths

def return_metrics(gt_path, prediction_path):
    with open(gt_path) as f:
        gt = set([i.strip('\n') for i in f.readlines() if len(i)])
    with open(prediction_path) as f:
        pred = set([i.strip('\n') for i in f.readlines() if len(i)])

    len_dataset = 20

    TP, TN, FP, FN = (0, ) * 4
    for i in pred:
        if i in gt:
            TP += 1
            gt.remove(i)
        else:
            FP += 1
    FN = len(gt)
    TN = len_dataset - TP - FP - FN

    accuracy = (TP + TN) / len_dataset
    recall = TP / (TP + FN)

    print(f'Acc: {accuracy:.2f}')
    print(f'Recall: {recall:.2f}')
````

````python
def find_matching_lego_pieces(target_image_path, source_dir, lower_threshold, upper_threshold):
    target_image = cv2.imread(target_image_path)
    target_image_gray = cv2.cvtColor(target_image, cv2.COLOR_BGR2GRAY)

    # Calculate Hu moments for the target image
    target_moments = cv2.moments(target_image_gray)
    target_hu_moments = cv2.HuMoments(target_moments)
    
    matching_images = set()

    for filename in os.listdir(source_dir):
        if filename.endswith(".jpg"):
            source_image_path = os.path.join(source_dir, filename)
            source_image = cv2.imread(source_image_path)
            source_image_gray = cv2.cvtColor(source_image, cv2.COLOR_BGR2GRAY)

            # Calculate Hu moments for the source image
            source_moments = cv2.moments(source_image_gray)
            source_hu_moments = cv2.HuMoments(source_moments)

            # Euclidean distance to compare Hu moments
            dist = np.linalg.norm(target_hu_moments - source_hu_moments)
            
            # print(f"{dist:.7f}", filename)
            # Set a threshold for matching
            if lower_threshold < dist and dist < upper_threshold:
                matching_images.add(filename)

    return matching_images
````

````python
target_image_path = Path(os.path.abspath('')) / 'Datasets' / 'test1' / 'target.jpg'
source_dir = Path(os.path.abspath('')) / 'Datasets' / 'test1' / 'source'

matching_images = find_matching_lego_pieces(target_image_path, source_dir, 0.000006, 0.00002)

print("Matching images test1:")
i = 0
for img in matching_images:
    print(f"{i}: {img}")
    i+=1

# Write results to a file
with open('test1.txt', 'w') as f:
   for filename in matching_images:
        f.write(filename + '\n')

# Calculate metrics
gt_test1 = Path(os.path.abspath('')) / 'Datasets' / 'gt' / 'test1.txt'
pred = Path(os.path.abspath('')) / 'test1.txt'
return_metrics(gt_test1, pred)
````

````python
target_image_path = Path(os.path.abspath('')) / 'Datasets' / 'test2' / 'target.jpg'
source_dir = Path(os.path.abspath('')) / 'Datasets' / 'test2' / 'source'

matching_images = find_matching_lego_pieces(target_image_path, source_dir, 0.000006, 0.000025)

print("Matching images test2:")
i = 0
for img in matching_images:
    print(f"{i}: {img}")
    i+=1

# Write results to a file
with open('test2.txt', 'w') as f:
   for filename in matching_images:
        f.write(filename + '\n')

# Calculate metrics
gt_test1 = Path(os.path.abspath('')) / 'Datasets' / 'gt' / 'test2.txt'
pred = Path(os.path.abspath('')) / 'test2.txt'
return_metrics(gt_test1, pred)
````

# Задача №2 - Найти шаблон на изображении с помощью HOG и кросс-корреляции

Ваша задача - написать программу, которая сможет найти шаблон одного глаза на нескольких изображениях, в том числе затемненных. Использовать нужно HOG представление изображений, поиск шаблона сделать с помощью кросс-корреляции 

Изображение глаз человека и шаблон, который нужно на нем найти, находятся в папке hog

````python
def process_image(image_path, template_path, threshold_lower, threshold_upper, title):
    image = cv2.imread(image_path)
    template = cv2.imread(template_path)

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)

    # Calculate HOG features
    fd, hog_image = hog(
        image,
        orientations=8,
        pixels_per_cell=(20, 20),
        cells_per_block=(1, 1),
        visualize=True,
        channel_axis=-1,
    )
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

    fd_template, hog_template = hog(
        template,
        orientations=8,
        pixels_per_cell=(20, 20),
        cells_per_block=(1, 1),
        visualize=True,
        channel_axis=-1,
    )
    hog_template_rescaled = exposure.rescale_intensity(hog_template, in_range=(0, 10))

    # Match template
    hog_image_rescaled = hog_image_rescaled.astype(np.float32)
    hog_template_rescaled = hog_template_rescaled.astype(np.float32)
    result = cv2.matchTemplate(hog_image_rescaled, hog_template_rescaled, cv2.TM_CCOEFF_NORMED)

    # Find matches within the thresholds
    loc = np.where((result <= threshold_upper) & (result >= threshold_lower))
    h, w, _ = template.shape

    for pt in zip(*loc[::-1]):
        cv2.rectangle(image, pt, (pt[0] + w, pt[1] + h), (0, 0, 255), 3)

    plot_transform_result(result, image, title)
    
template_path = 'hog/eye.png'
process_image('hog/eyes.jpg', template_path, 0.70, 0.7125, 'Eyes')
process_image('hog/eyes_dark.jpg', template_path, 0.70, 0.76, 'Eyes in Dark')
process_image('hog/woman.jpg', template_path, 0.62575, 0.626, 'Woman') 
````
